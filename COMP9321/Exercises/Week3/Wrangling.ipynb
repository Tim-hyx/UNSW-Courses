```python
def get_year(all_CVE):
    lst = []
    for c in all_CVE.split("-"):
        lst.append(c)
    return lst[1]

df["CVE-Year"] = df["CVE"].apply(get_year) 
df.head()
```

```python
Our_grouped_CVEs= df.groupby('CVE-Year') ['CVE'].count()
Our_grouped_CVEs
```
